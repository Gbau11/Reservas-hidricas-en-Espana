```python
esp_gb = esp.groupby('FECHA').agg({'AGUA_ACTUAL': 'sum', 'AGUA_TOTAL': 'sum'})
```

```python
esp_gb = esp_gb.iloc[2:]
esp_gb
```

```python
esp_gb['capacidad'] = esp_gb['AGUA_ACTUAL'] * 100 / esp_gb['AGUA_TOTAL']
```

```python
esp_gb.capacidad.describe()
```

```python
esp_gb.info()
```

```python
esp_gb
```

```python
esp_gb.reset_index(drop=False, inplace=True)
```

```python
esp_gb
```

```python
esp_gb['year'] = esp_gb['FECHA'].dt.year

max_cap = esp_gb.groupby('year')['capacidad'].max()

max_cap
```

```python
def max_capacity_filter(x):
    return x[x['capacidad'] == x['capacidad'].max()]

result = esp_gb.groupby('year').apply(max_capacity_filter)

result.reset_index(drop=True, inplace=True)

result
```

```python
def min_capacity_filter(x):
    return x[x['capacidad'] == x['capacidad'].min()]

result2 = esp_gb.groupby('year').apply(min_capacity_filter)

result2.reset_index(drop=True, inplace=True)

result2
```

```python
result2.capacidad.mean()
```

```python
result.capacidad.mean()
```

```python
result = result.rename(columns={'capacidad':'max_cap'})
```

```python
result2 = result2.rename(columns={'capacidad':'min_cap'})
```

```python
result = pd.merge(result,result2, how='inner',on='year')
```

```python
url = 'https://datosmacro.expansion.com/demografia/poblacion/espana'
```

```python
pop_esp = pd.read_html(url)[0]
```

```python
pop_esp = pop_esp.iloc[::2]
```

```python
pop_esp.reset_index(drop=True,inplace=True)
pop_esp['Fecha'] = pop_esp['Fecha'].str.replace('Diciembre','')
pop_esp['Fecha']= pop_esp['Fecha'].astype(int)
pop_esp.drop(columns=['Densidad','Hombres','Mujeres'], inplace=True)
pop_esp.rename(columns={'Fecha':'year'},inplace=True)
pop_esp['Población']= pop_esp['Población'].str.replace('.','')
pop_esp['Población']= pop_esp['Población'].astype(int)
```
